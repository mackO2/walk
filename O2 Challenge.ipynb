```python
# import python libraries
import pandas as pd
```

```python
#import the csv file
#only import the Name, Distance and Units data from the data.csv file as a pandas dataframe
df = pd.read_csv('data2.csv')
```

```python
# just use a few columns
df_ = df[['Name','Distance','Units']]
df_
```

```python
# Add a 'miles' column (dataseries) to the dataframe and set each row value for 'Miles' to 0
df_.loc[:,'Miles'] = 0
df_
```

```python
# Convert distance and record value in Miles column; df_.loc[rows_by_mask,col_is_Miles] = df_['Distance'] * conversion
# miles
df_.loc[(df_['Units'] == 'Miles'), 'Miles'] = df_['Distance'] * 1
# Kilometers
# 0.621 miles = 1km
df_.loc[(df_['Units'] == 'Km'), 'Miles'] = df_['Distance'] * 0.621
# steps
# 2000 steps = 1m
df_.loc[ (df_['Units'] == 'Steps'), 'Miles'] = df_['Distance'] * 0.0005
# round to 2 decimal places
df_ = df_.round(2)
df_
```
